## Ansatze for Quantum State Preparation

For an N-dimensional datapoint $$A = (a_0, a_1, \dots, a_{N-1})$$ with $\sum_{k}|a_k|^2 = 1$, we must prepare the target quantum state $$ |\psi\rangle = \sum_{k=0}^{N-1}a_k|k\rangle $$ from an initial state of $|0\rangle^{\otimes n}$, where $N = 2^n$


```python
from pqc import *
import numpy as np
```

```python
import matplotlib.pyplot as pltb
```

With `initialize`

```python
costs = []
times = []
for N in range(2, 65):
    A_vals = list(np.random.uniform(-1, 1, N))

    diff = int(2**np.ceil(np.log2(N)) - N)
    for _ in range(diff):
        A_vals.append(0)

    A_vals = np.divide(A_vals, np.sqrt(sum(np.square(A_vals))))

    print(f"N = {N}", end="\t")
    # print(f"Sum of probs : {sum(np.square(A_vals))}")

    # print(f"Target : {A_vals}")

    sv, cost, time = init_load(A_vals)

    # print(f"Statevector : {sv}")
    print(f"Cost : {cost}\tTime : {time}")
    # print(f"Execution time : {time}s")

    costs.append(cost)
    times.append(time)

e1 = open('expt1_real/desc.txt', 'w')
e1.write('reps = ceil(N/2), opt = BFGS \n')
e1.close()

e1 = open('expt1_real/cost.txt', 'w')
e1.write(str(costs))
e1.close()

e1 = open('expt1_real/time.txt', 'w')
e1.write(str(times))
e1.close()
```
